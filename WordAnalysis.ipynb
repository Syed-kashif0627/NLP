```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
# import matplotlib.pyplot as plt

def wordAnalysis(text):

    words = word_tokenize(text)

    stopWords = set(stopwords.words('english'))

    filterWords = [word.lower() for word in words if word.lower() not in stopWords]

    freqDist = FreqDist(filterWords)

    print("Word Frequency:")
    for word, cnt in freqDist.items():
        print(word, ":", cnt)

    print("\nPlotting frequency distribution...")
    freqDist.plot(30, cumulative=False)  # Adjust the number of words to plot as needed

# # Download necessary resources
# nltk.download('stopwords')
# nltk.download('punkt')

# Input sentence from the user
sentence = input("Enter sentence to perform word analysis: ")
print(f"\nWord analysis on the sentence \"{sentence}\" is:")
wordAnalysis(sentence)
```
